```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/Tennis.csv')

# Display the first few rows and information about the dataset
print("Dataset head:")
display(df.head())
print("\nDataset Info:")
df.info()
```

The dataset contains categorical features. I will convert these into numerical representations using one-hot encoding, which is suitable for decision trees and avoids implying ordinal relationships.

```python
# Identify categorical columns (excluding the target 'play tennis')
categorical_cols = ['Outlook', 'Temprature', 'humidity', 'windy']

# Apply one-hot encoding
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Separate features (X) and target (y)
X = df_encoded.drop('play tennis', axis=1)
y = df_encoded['play tennis'].apply(lambda x: 1 if x == 'yes' else 0) # Convert 'yes'/'no' to 1/0

print("Encoded features (X) head:")
display(X.head())
print("\nTarget (y) head:")
display(y.head())
```

Now that the data is prepared, I will build and train a Decision Tree Classifier using the `entropy` criterion and then visualize the resulting tree.

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Initialize the Decision Tree Classifier with entropy criterion
dtree = DecisionTreeClassifier(criterion='entropy', random_state=42)

# Train the model
dtree.fit(X, y)

# Visualize the Decision Tree
plt.figure(figsize=(15, 10))
plot_tree(dtree, filled=True, feature_names=X.columns.tolist(), class_names=['No', 'Yes'], rounded=True)
plt.title("Decision Tree Classifier (Criterion: Entropy)")
plt.show()
```
